Q 17.What are point estimates and interval estimates in statistics

  - 1.`Point Estimate

A point estimate is a single value that serves as the best guess of a population parameter.



- 2. Interval Estimate

An interval estimate provides a range of values within which the population parameter is expected to lie, with a certain level of confidence.



Q 18.What is the significance of confidence intervals in statistical analysis
1. Estimate Precision

CIs give not just a single estimate (point estimate), but a range.

Example: Instead of saying "the average height is 170 cm", a 95% CI might be "170  3 cm", or (167 cm, 173 cm), showing uncertainty around the estimate.

2. Inference about Population Parameters
Based on sample data, CIs allow us to infer about the entire population without measuring every individual.

3. Statistical Significance Testing
If a CI for a difference in means (like treatment vs. placebo) does not include 0, the difference is statistically significant.

Example: A 95% CI for a mean difference is (2.1, 5.7)  likely real effect.

4. Helps Avoid Overconfidence
Reporting just a point estimate can be misleading.

CIs remind us that sample statistics are subject to sampling variability.

5. Comparison Across Groups
When comparing two or more groups, overlapping or non-overlapping confidence intervals can indicate if differences are likely meaningful.

6. Better Communication of Uncertainty
CIs make results more transparent and are widely used in research, medicine, economics, and polling.

Q 21.What are the assumptions for applying the Central Limit Theorem

 1. Random Sampling

The data must be collected through random sampling.

Each sample should be independent of the others.

This ensures that the sample is representative of the population.

2. Sample Size

The sample size should be "sufficiently large":

Usually n  30 is considered adequate.

If the population distribution is strongly skewed or non-normal, a larger sample may be needed.

If the population is already normally distributed, even a small sample size is fine.

3. Independence of Observations

Each data point in the sample should be independent of the others.

This means the outcome of one observation should not affect the outcome of another.

4. Finite Variance

The population must have a finite variance (i.e., standard deviation exists and is not infinite).

CLT doesn't work well if the population has extreme outliers or heavy tails (like in Cauchy distributions).
